Poner Marca en el Mapa

```python
from folium.plugins import MiniMap
from branca.element import Figure

popuptext= "<b>Punto Importante de la UNA<b>"

my_map = folium.Map(location=[9.97071052359151, -84.1291289162911],zoom_start=16,tiles="stamenterrain")

#ESCINF

folium.Marker(location=[9.97071052359151, -84.1291289162911],popup=popuptext,icon=folium.Icon(color="orange",icon="ok-sign")).add_to(my_map)

folium.Circle(location=[9.97071052359151, -84.1291289162911],color="purple",weight=6,fill_color="red",fill_opacity=0.5,tooltip="ESCINF").add_to(my_map)

#Centro Internacional de Pol√≠tica Econ√≥mica para el Desarrollo Sostenible

folium.Marker(location=[9.971343409685947, -84.12717219251734],popup=popuptext,icon=folium.Icon(color="orange",icon="ok-sign")).add_to(my_map)

folium.Circle(location=[9.971343409685947, -84.12717219251734],color="purple",weight=6,fill_color="red",fill_opacity=0.5,tooltip="Centro Internacional de Pol√≠tica Econ√≥mica para el Desarrollo Sostenible").add_to(my_map)

#Escuela de Medicina Veterinaria

folium.Marker(location=[9.977920641389941, -84.12744186651554],popup=popuptext,icon=folium.Icon(color="orange",icon="ok-sign")).add_to(my_map)

folium.Circle(location=[9.977920641389941, -84.12744186651554],color="purple",weight=6,fill_color="red",fill_opacity=0.5,tooltip="Escuela de Medicina Veterinaria").add_to(my_map)

#Centro de Investigaciones Ap√≠colas Tropicales

folium.Marker(location=[9.972928571976162, -84.12957953032281],popup=popuptext,icon=folium.Icon(color="orange",icon="ok-sign")).add_to(my_map)

folium.Circle(location=[9.972928571976162, -84.12957953032281],color="purple",weight=6,fill_color="red",fill_opacity=0.5,tooltip="Centro de Investigaciones Ap√≠colas Tropicales").add_to(my_map)

#Residencias Estudiantiles Benjam√≠n Nu√±ez 2

folium.Marker(location=[9.969521287942056, -84.12808250014724],popup=popuptext,icon=folium.Icon(color="orange",icon="ok-sign")).add_to(my_map)

folium.Circle(location=[9.969521287942056, -84.12808250014724],color="purple",weight=6,fill_color="red",fill_opacity=0.5,tooltip="Residencias Estudiantiles Benjam√≠n Nu√±ez 2").add_to(my_map)

minimap=MiniMap(tile_layer="stamenterrain",position="bottonright")

my_map.add_child(minimap)

my_map.save("C:Folium.html")
my_map
```
